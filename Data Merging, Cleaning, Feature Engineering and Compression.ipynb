### **Importing Libraries and Dataset**

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import pyarrow.csv as pv
import pyarrow.parquet as pq
```

```python
# Importing the October 2019 dataset
data1 = pv.read_csv('2019-Oct.csv')
```

```python
# Sending the data from pyarrow to pandas dataframe
data1_p = data1.to_pandas()
```

```python
# Fetching first 5 rows of the dataset 2019 October
data1_p.head()
```

```python
# Importing the November 2019 dataset
data2 = pv.read_csv('2019-Nov.csv')
```

```python
# sending the data from pyarrow to pandas dataframe
data2_p = data2.to_pandas()
```

```python
# Fetching first 5 rows of the dataset 2019 November
data2_p.head()
```

```python
# Merging the two datasets
df = pd.concat([data1_p, data2_p], axis=0)
```

```python
# Getting the shape of the merged dataset
df.shape
```

```python
# Fetching first 5 rows of the merged dataset
df.head()
```

```python
# Checking the memory usage of the dataframe
df.memory_usage(deep=True).sort_values(ascending=False)
```

```python
# Dropping the user_session column as it is not required for analysis
df = df.drop(columns=['user_session'])
```

```python
df.head()
```

```python
# Convert Object type columns to category type to reduce memory usage
df['event_type']= df['event_type'].astype('category')
df['brand']=df['brand'].astype('category')
df['category_code']=df['category_code'].astype('category')
```

```python
# Converting the 'event_time' column to datetime format
df['event_time'] = pd.to_datetime(df['event_time'])
```

```python
# Downcasting the numerical columns to reduce memory usage
df['product_id'] = df['product_id'].astype('int32')
df['category_id'] = df['category_id'].astype('int32')
df['user_id'] = df['user_id'].astype('int32')
df['price'] = df['price'].astype('float32')
```

```python
# Extracting year, month and day from the 'event_time' column
df['year'] = df['event_time'].dt.year
df['month'] = df['event_time'].dt.month
df['day'] = df['event_time'].dt.day
```

```python
# Checking for null values in the dataset
df.isnull().sum()
```

```python
# Saving the cleaned and feature engineered dataset in parquet format with zstd compression
df.to_parquet(
    "final_compressed.parquet",
    index=False,
    compression="zstd"
)
```
